### Merkmale
- Primary_Label = Zielvogel -> Bezeichnung: Aus common_name kombiniert. Zebra Dove -> zebdov
- Secondary_Label = Vögel im Hintergrund -> Bezeichnung: Aus common_name kombiniert. Zebra Dove -> zebdov
- type = Eigenschaften der Stimme
- latitude = Breitegrad
- longitude = Längengrad
- scientific_name = Wissenschaftlicher Name der Vogelart
- common_name = Allgemeine Bezeichnung der Vogelart
- author = Person, welche die Aufnahme durchgeführt hat (Offensichtlich unwichtig für das Model)
- license = Lizenz der Aufnahme (Offensichtlich unwichtig für das Model)
- rating = Aufnahmequalität (siehe: Xeno-Canto: Rating Bedeutung)
- time = Tageszeit zu der die Aufnahmen stattgefunden hat
- url = Pfad zur Quelle woher die Audio-Datei stammt (Offensichtlich unwichtig für das Model)
- file_name = Pfad zur Audio-Datei 

## Informationen zu den Merkmalen

### Details: Rating
Nach der Begutachtung mehrerer Aufnahmen sind folgende Aspekte zu berücksichtigen:
- A entspricht 5.0
- E entspricht dem kleinsten möglichen Wert (vermutlich 0.0)

##### Xeno-Canto: Rating Bedeutung
Gibt es irgendwelche Richtlinien für die Bewertung der Aufnahmen?

Benutze die folgenden allgemeinen Richtlinien bei der Bewertung von Aufnahmen auf Xeno-Canto. Bewertungen sind offenbar subjektiv und werden unvermeidlich zwischen verschiedenen Personen etwas schwanken, aber diese Richtlinien sollten die Beständigkeit erhöhen.

    A: Laut und klar
    B: Klar, doch ein wenig entfernt, oder einige Störungen durch andere Schallquellen
    C: Einigermaßen klar oder deutliche Störungen
    D: Sehr leise Aufnahme oder starke Störungen
    E: Kaum hörbar
Quelle: FAQ - https://xeno-canto.org/help/FAQ
