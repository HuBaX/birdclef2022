## Informationen zu den Merkmalen

### Details: Rating
Nach der Begutachtung mehrerer Aufnahmen sind folgende Aspekte zu berücksichtigen:
- A entspricht 5.0
- E entspricht dem kleinsten möglichen Wert (vermutlich 0.0)

##### Xeno-Canto: Rating Bedeutung
Gibt es irgendwelche Richtlinien für die Bewertung der Aufnahmen?

Benutze die folgenden allgemeinen Richtlinien bei der Bewertung von Aufnahmen auf Xeno-Canto. Bewertungen sind offenbar subjektiv und werden unvermeidlich zwischen verschiedenen Personen etwas schwanken, aber diese Richtlinien sollten die Beständigkeit erhöhen.

    A: Laut und klar
    B: Klar, doch ein wenig entfernt, oder einige Störungen durch andere Schallquellen
    C: Einigermaßen klar oder deutliche Störungen
    D: Sehr leise Aufnahme oder starke Störungen
    E: Kaum hörbar
Quelle: FAQ - https://xeno-canto.org/help/FAQ

### Distribution for unique primary_labels

```python
plt.figure(figsize=(30,15))

for index, column in enumerate(['rating','primary_label']):
    ax = plt.subplot(3,5, index+1)
    dataset_metadata[column].value_counts().plot(ax=ax, kind="bar")
    ax.set_title(column)

plt.subplots_adjust(hspace=0.6)
plt.show()
```

Since the primary_label has many unique entries (152), we'll try to split up the results.

```python
plt.figure(figsize=(30,15))
viewsize = 15
for i in range(int(len(dataset_metadata['primary_label'].unique())/viewsize + 1)):
    ax = plt.subplot(3,5, i+1)
    start = i * viewsize
    end = (i+1) * viewsize
    dataset_metadata['primary_label'].value_counts()[start:end].plot(ax=ax, kind="bar")
    ax.set_title(f'primary_label:{start}, {end}')
    
plt.subplots_adjust(hspace=0.6)
plt.show()

```

Collect all entries with value_count below the mean

```python
primary_label_value_count_mean = dataset_metadata['primary_label'].value_counts().mean()

# How many primary_label are below mean
primary_label_below_mean_list = list(dataset_metadata['primary_label'].value_counts()[dataset_metadata['primary_label'].value_counts() < primary_label_value_count_mean].index)
print(f"Below mean: {len(dataset_metadata['primary_label'].value_counts()[dataset_metadata['primary_label'].value_counts() < primary_label_value_count_mean])}")
print(f"All entries: {len(dataset_metadata['primary_label'].value_counts())}")
print(primary_label_below_mean_list)
```

Connect Labels with rating 

```python
primary_label_below_mean_dataset = dataset_metadata[dataset_metadata['primary_label'].isin(primary_label_below_mean_list)]
primary_label_below_mean_dataset_set = primary_label_below_mean_dataset[['primary_label', 'rating']]

plt.figure(figsize=(30,20))

for index, label in enumerate(primary_label_below_mean_list):
    #print(f"{label} : {primary_label_below_mean_dataset_set.where(primary_label_below_mean_dataset_set['primary_label'] == label).dropna()['rating'].value_counts()}")
    ax = plt.subplot(20,7,index+1)
    primary_label_below_mean_dataset_set.where(primary_label_below_mean_dataset_set['primary_label'] == label).dropna()['rating'].value_counts().plot(ax=ax, kind='bar')
    ax.set_title(label)
plt.subplots_adjust(hspace=3.0)
plt.show()
```

As Dict:

```python
for index, label in enumerate(primary_label_below_mean_list):
    print(f"{label} : {primary_label_below_mean_dataset_set.where(primary_label_below_mean_dataset_set['primary_label'] == label).dropna()['rating'].value_counts().to_dict()}")
```

```python
for index, label in enumerate(primary_label_below_mean_list):
    print(f"{label} : \n{primary_label_below_mean_dataset_set.where(primary_label_below_mean_dataset_set['primary_label'] == label).dropna()['rating'].value_counts().sort_index()}")
```

```python
for index, label in enumerate(primary_label_below_mean_list):
    label_value_count = primary_label_below_mean_dataset_set.where(primary_label_below_mean_dataset_set['primary_label'] == label).dropna()['rating'].value_counts()
    if label_value_count.sum() < 10:
        print(f"{label} : {label_value_count.to_dict()}")
        
```
