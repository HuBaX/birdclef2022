```python
sns.set(font_scale=2)

for run in os.listdir(LOG_PATH):
    metrics = pd.read_csv(f'{LOG_PATH}/{run}/metrics.csv')
    del metrics["epoch"]
    metrics.set_index("step", inplace=True)

    plt.figure(figsize=(5,5))
    rel = sns.relplot(data=metrics, kind="line", height=12, aspect=1.7)
    (rel.map(plt.axhline, y=1, color="0.7", dashes=(2, 1), zorder=0)).set_titles(f"{run}")
    plt.title(f"{run}", fontdict={'fontsize': 20})
    plt.grid()
```
